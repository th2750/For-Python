```python
import pandas as pd
data = pd.read_csv('Desktop/python-novice-gapminder-data/data/gapminder_gdp_americas.csv', index_col='country')

data = data.drop(labels='continent', axis=1)
data

#exract the year from last four charts of each column
years = data.columns.str.strip('gdpPercap_')
#convert years to ints
data.columns = years.astype(int)

plt.style.use('ggplot')

chileplot = data.loc['Chile'].plot(label='Chile')
mexicoplot = data.loc['Mexico'].plot(label='Mexico')

plt.xlabel('Year')
plt.ylabel('GDP Per Capita (in millions USD)')
plt.legend(loc='upper left')
plt.title('GDP Growth')
plt.savefig('GDP Growth Mexico Chile')
```
